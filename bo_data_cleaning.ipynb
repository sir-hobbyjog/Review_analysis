```python
import pandas as pd

df1 = pd.read_csv('rt_reviews_cleaned.csv')
df2 = pd.read_csv('rotten_tomatoes_movies.csv')
```

```python
df1.head()
```

```python
df2.head()
```

```python
# Create a dictionary that assigns each unique rotten_tomatoes_link to the corresponding unique movie_title
movie_title_dict = df2.set_index('rotten_tomatoes_link')['movie_title'].to_dict()

# Create a new column for df1 titled 'movie_title'
df1['movie_title'] = ''
```

```python
print(movie_title_dict)
```

```python
df1.head() 
```

```python
df1["movie_title"] = df1["rotten_tomatoes_link"].map(movie_title_dict)
```

```python
# Save new DataFrame to CSV
df1.to_csv('rt_reviews_cleaned_with_names.csv', index=False)
```
